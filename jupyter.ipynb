## Jupyter Notebook を Jupyter 上実行するには
Jupyter Notebook ファイルは、単独では Jupyter として実行することができません。Notebook ファイルを Jupyter 上で動かすことで、Python コードなどが実行され、その結果を Notebook ファイルに記録することができます。

この Notebook ファイルを Jupyter 上で実行するには以下の2つの方法があります。
1. Github 上に置かれたこのファイルのページ最上部に表示されている「Open in Colab」のリンクをクリックすることで、クラウド仮想環境の Google Colaboratory（以下、Colab）上で Notebook ファイルを Jupyter を使って開くことができます。
1. また、このファイル上部にある「Raw」を右クリックして「リンク先のファイルを別名で保存」を選択することで、このファイル（jupyter.ipynb）をダウンロードすることができます。ダウンロードしたファイルを Jupyter Notebook として開くことで、Jupyter 上で Notebook ファイルを自分のPC上で実行することができます。

2つ目の方法で Notebook ファイルを実行するためには、自分のPC上に Jupyter をインストールする必要があります。科目のページに「VS Code で Jupyter 環境を構築」する方法が開設されています。

## Jupyter Notebook を実行する
Python のコードは以下のようなBOX（コードセルと呼ばれます）に記述され、このセルを選択した状態で Shift+Enter または、セル左横の ▷ マークをクリックすることでセル内の Python コードを実行することができます。

以下の2つのセルを上から順に実行してみましょう。

```python
a = 1234
b = 5678
```

```python
b/a
```

b/a の計算結果がセルの下に表示されたと思います。
次に、a, b の値を変更するなどセル内のコードを書き換えてから、再び2つのセルを実行してみてください。b/a の結果の値が変わっているはずです。
このように、Jupyter Notebook ではコードセル内の Python コードを対話的に修正、実行することができます。

Jupyter Notebook のコードセルは、ノートブック内に複数記述することができます。ノートブック内のコードセルは独立しておらず、名前空間を共有しています。
したがって、以下のセルを実行した後、再び上にある b/a のセルを評価すると、a, b の値は直近に評価された値 $a=0.1, b=123^{99}$にもとづいて計算されることになります。


```python
a = 0.1
b = 123**99
```

同じノートブック内のセルでも、セルが評価される順番によって結果が異なることに注意してください。そのセルが何番目に実行されたかは、セルの左横（▷の位置）に表示されます。
